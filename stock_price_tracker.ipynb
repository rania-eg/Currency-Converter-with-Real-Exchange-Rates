```python
import yfinance as yf
```

```python
import pandas as pd
```

```python
symbol = input("Enter stock symbol (e.g. AAPL, TSLA, GOOG): ").upper()
```

```python
data=yf.download(symbol, period="7d", interval="1d")
```

```python
data["Daily Change (%)"] = data["Close"].pct_change() * 100
```

```python
latest = data.tail(1)
print("\nðŸ“Š Latest Stock Data:\n")
print(latest[["Close", "Daily Change (%)"]])
```

```python
data["Close"].plot(title=f"{symbol} Stock Price (Last 7 Days)", ylabel="Price ($)", grid=True)
```
